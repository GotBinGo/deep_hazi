```python
import requests
import re
import numpy as np

url = 'http://idojarasbudapest.hu/archivalt-idojaras'

honapok = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
hosszok = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
evek = ["2016", "2017"]

#honapok = ['01', '02']
evek = ['2012', '2013', '2014', '2015', '2016', '2017']

data = np.array([]).reshape(-1, 4)


for ev in evek:
    nap = 0
    for i, ho in enumerate(honapok):
        print(ev+' '+ho)
        payload = {'ev':ev, 'ho':ho, 'button':'Mehet'}
        r = requests.get(url, params=payload)

        r = requests.post(url, data=payload)
        r = r.content.decode("utf-8")

        a = re.findall('<td .*?</td>', r)
        a = list(map(lambda a: a.split('>')[1].split('<')[0], a))
        a = np.array(a).reshape(-1, 3)
        print(len(a), 'lett')
        if(len(a) > hosszok[i]): #szokoev levagas
            a = a[0:hosszok[i]] 

        if(len(a) < hosszok[i]): #ha hianyzik duplazas 
            if (hosszok[i]-len(a)) < 3:
                print('POTOLVA',hosszok[i]-len(a))
                a = np.append(a,a[-(hosszok[i]-len(a)):], axis=0)
                
            else:
                print('HIANYOS HONAP')
                  
       
        print(hosszok[i], 'lenne')
        print('\n')
        b = np.arange(nap+1,nap+len(a)+1).reshape(-1,1)
        nap = nap+len(b)
        c = np.append(b, a, axis=1)

        #print(c)
        data = np.append(data, c)



#print(r)
data = data.reshape(-1, 4)
```
